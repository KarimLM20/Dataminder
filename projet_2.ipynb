```python
import pandas as pd
import ast
from collections import defaultdict

# 1. Nettoyage des 'characters' : transformer les chaînes de liste en vraies listes
def safe_eval(x):
    try:
        return ast.literal_eval(x) if isinstance(x, str) and x.startswith('[') else [x]
    except:
        return [x]

merged_df['characters'] = merged_df['characters'].apply(safe_eval)

# 2. Regrouper les rôles par film (tconst)
def regrouper_par_role(group):
    role_dict = defaultdict(list)
    for _, row in group.iterrows():
        role = row['category']
        noms = row['characters']
        if isinstance(noms, list):
            role_dict[role].extend(noms)
    return dict(role_dict)

roles_df = merged_df.groupby('tconst').apply(regrouper_par_role).reset_index()
roles_df.columns = ['tconst', 'personnes_par_rôle']

# 3. Sélectionner les colonnes à garder
colonnes_utiles = [
    'Unnamed: 0', 'tconst', 'titleType', 'primaryTitle', 'originalTitle',
    'isAdult', 'startYear', 'endYear', 'runtimeMinutes', 'genres_x',
    'video', 'vote_average', 'vote_count',
    'production_companies_name', 'production_companies_country',
    'ordering', 'nconst'
]

df_selection = merged_df[colonnes_utiles].drop_duplicates(subset='tconst')

# 4. Fusionner avec la colonne "personnes_par_rôle"
final_df = pd.merge(df_selection, roles_df, on='tconst', how='left')

# 5. Exporter
final_df.to_csv('C:/Users/karbe/KarimLM20/Dataminder/films_allégés_avec_roles.csv', index=False, encoding='utf-8')

# 6. Afficher un aperçu
print(final_df.head())
```
